In this notebook, we make an interesting observation about dealing with sines and cosines of discrete data. Let's begin by plotting

$g(t) = 5 \sin \left(\frac{2 \pi t }{ T } \right)$, with $T=0.5$

To make this look like a nice continuous function, we typically pick a pretty fine grid of $t$ values to plot. Let's go from $t=0$ to $t=1$, spread over $129$ points.

```python
t = np.linspace(0., 1.0, 129)
g_fct = 5. * np.sin(2. * np.pi * t / 0.5)
```

<b> Q2. Now do the same (with 33 points along the $t$ axis) but for $T=1/66\approx 0.015$. Plot the new set of points on top of the old ones, but in black and using "x" markers instead of circles. Think about what you expect this to look like, and then make the plots. Warning: counterintuitive result coming up!

<b> Solution:

```python
g2_fct = 5. * np.sin(2. * np.pi * t / (1/66.))
```

```python
f, ax = plt.subplots(figsize=(9,6))
ax.plot(t,g_fct,lw=2,c='red',marker='o')
ax.plot(t,g2_fct,lw=2,c='black',marker='x')
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$g(t)$")
```

Bizarre! We would expect the new curve to have a much shorter period, but when we plot them together, they're somehow exactly the same?! What's going on?

<b> Q3. Make sense of what's going on by repeating the last problem but with $129$ points along the $t$ axis. Can you explain what happened?

<b> Solution:

```python
t = np.linspace(0., 1.0, 129)
g_fct = 5. * np.sin(2. * np.pi * t / 0.5)
g2_fct = 5. * np.sin(2. * np.pi * t / (1/66.))
```

```python
f, ax = plt.subplots(figsize=(9,6))
ax.plot(t,g_fct,lw=2,c='red',marker='o')
ax.plot(t,g2_fct,lw=2,c='black',marker='x')
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$g(t)$")
```

The continuous versions of those two functions were different, as one can see on the last plot. But when we sample the $t$ axis coarsely, the two sines are completely indistinguishable!

As one implication of this, think back to our Fourier Series. When we were trying to express a *continuous* function as a sum of sines and cosines, we needed an *infinite* series of them. But suppose we didn't have a continuous function of $t$, but just had $33$ datapoints along the $t$ axis. Would we need both the red sine curve and the black sine curve? No! Because they're indisinguishable anyway. Generalizing this result, it turns out that if we have a discrete set of points, we can express them in terms of a *finite* series of sines and cosines.

This turns out to be part of a much deeper topic. To find out more, google "aliasing" or the "Nyquist-Shannon Sampling Theorem".
