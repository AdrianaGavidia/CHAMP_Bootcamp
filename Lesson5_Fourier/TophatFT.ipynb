```python
import numpy as np
%pylab inline
```

```python
def gt(t,a):
    out = t.copy()
    for i,elem in enumerate(t):
        if abs(elem) > a:
            out[i] = 0.
        else:
            out[i] = 1.
    return out

def gtilde(f,a):
    return a* np.sinc(f * a)
```

```python
t = np.linspace(-10.,10.,1000)
```

```python
f, ax = plt.subplots(figsize=(9,6))
ax.plot(t,gt(t,1.),lw=2,c='blue')
ax.plot(t,gt(t,2.5),lw=2,c='red')
ax.plot(t,gt(t,5.),lw=2,c='black')
ax.axhline(0, color='black',ls='--')
ax.axvline(0, color='black',ls='--')
ax.set_ylim(-0.2, 1.2)
```

```python
freq = np.linspace(-10.,10.,1000)
```

```python
f, ax = plt.subplots(figsize=(9,6))
ax.plot(freq,gtilde(freq,1.),lw=2,c='blue')
ax.plot(freq,gtilde(freq,2.5),lw=2,c='red')
ax.plot(freq,gtilde(freq,5.),lw=2,c='black')
ax.axhline(0, color='black',ls='--')
ax.axvline(0, color='black',ls='--')
ax.set_xlim(-5.,5.)
```
