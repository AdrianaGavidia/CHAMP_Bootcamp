# Comparing Pre/Post Survey Results

```python
from textwrap import wrap
p.figure(figsize=(20,10))
xs = n.arange(1,18,1)
colors = p.cm.rainbow(n.linspace(0,1,len(xs)))
labels = []
befores = []
afters = []
for i in xs:
    before = n.mean(Ai[i])
    after = n.mean(Af[i])
    befores.append(before)
    afters.append(after)
    labels.append(Q[i])
    
befores = n.array(befores)
afters = n.array(afters)
labels = n.array(labels)

diff = afters-befores
order = n.argsort(diff)

befores = befores[order]
afters = afters[order]
labels = labels[order]

for i,c in zip(xs,colors):
    p.plot(i,befores[i-1], '.', color=c, ms=15, label="Pre-Summer" if i==1 else "")
    p.plot(i,afters[i-1], '*', color=c, ms=15, label="Post-Summer" if i==1 else "")
p.ylim(2.5,5.1)
p.legend(loc='lower right',prop={'size':16})
p.grid()
p.title("CHAMP Survey Pre vs. Post Summer", fontsize=16)
p.ylabel("Mean Score (5 = strongly agree)", fontsize=12)
p.xticks(xs, labels, rotation=45)
p.show()
```
